```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()             # Create a figure containing a single Axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the Axes.
plt.show()                           # Show the figure.
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) Axes.
plt.plot(x, x**2, label='quadratic',linewidth=4, linestyle=':')  # etc.
plt.plot(x, x**3, label='cubic', color="magenta",linewidth=3, linestyle='--')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.grid()
```

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)

plt.figure(layout='constrained')
plt.hist(x, 100, label="fdasfadsf")
plt.xlim([0, 200])
plt.grid([True])
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7), layout="constrained")
sx = np.random.rand(120)
sy = np.random.rand(120)
ax.scatter(sx, sy, s=30, facecolor='C0', edgecolor='k', alpha=0.5)
```

```python
# make data
x = np.linspace(0, 10, 100)
y = 4 + 1 * np.sin(2 * x)
x2 = np.linspace(0, 10, 25)
y2 = 4 + 1 * np.sin(2 * x2)

# plot
fig, ax = plt.subplots()

ax.plot(x2, y2 + 2.5, 's', markeredgewidth=2)
ax.plot(x, y, linewidth=2.0)
ax.plot(x2, y2 - 2.5, 'o-', linewidth=2)
```

```python
# make data:
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white")
```

```python
from matplotlib import cm # use colormap from matplot
# make data
x = np.arange(0, 10, 2)
ay = [1, 1.25, 2, 2.75, 3]
by = [1, 1, 1, 1, 1]
cy = [2, 1, 2, 1, 2]
y = np.vstack([ay, by, cy])

# plot
fig, ax = plt.subplots()
colors = plt.cm.copper(np.linspace(0.3, 1.0, len(x)))
ax.stackplot(x, y, colors=colors)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

# plot:
fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.show()
```

```python
x = [1, 2, 3, 4]
# plot
colors = plt.cm.hot(np.linspace(0.2, 0.9, len(x)))

fig, ax = plt.subplots(figsize=(3,3))
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 7)

# plot
fig, [ax1, ax2] = plt.subplots(1,2,figsize=(6,6))

ax1.contour(X, Y, Z) # plot1 no fill
ax2.contourf(X, Y, Z) #plot2 fill

ax1.set_aspect("equal")
ax2.set_aspect("equal")

plt.show()
```

```python
# Make data
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(4,4))
ax.scatter(xs, ys, zs)

plt.show()
```

```python
# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(4,4))
ax.plot_surface(X, Y, Z, cmap=plt.cm.twilight)

plt.show()
```
