```python
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
import numpy as np
```

```python
diabetes = load_diabetes(scaled=False)
print(diabetes.DESCR)
```

```python
print(diabetes.data.shape)
print(diabetes.feature_names)
print(diabetes.data)
```

```python
print(diabetes.target.shape)
print(diabetes.target)
```

```python
input_unscaled = diabetes.data
type(input_unscaled)
output_unscaled = diabetes.target
```

```python
age = input_unscaled[:,0]
sex = input_unscaled[:,1]
bmi = input_unscaled[:,2]
bp = input_unscaled[:,3]
disease = output_unscaled
```

```python
fig, axs = plt.subplots(2, 2, layout='constrained')
page, psex = axs[0, 0], axs[0, 1]
pbmi, pbp = axs[1, 0], axs[1, 1]

page.set_xlabel("Age (Years)")
psex.set_xlabel("Sex")
pbmi.set_xlabel("Body Mass Index")
pbp.set_xlabel("Average Blood Pressure")

page.set_ylabel("Disease Progression")
psex.set_ylabel("Disease Progression")
pbmi.set_ylabel("Disease Progression")
pbp.set_ylabel("Disease Progression")

page.scatter(age, output_unscaled,color="r",s=10)
psex.scatter(sex, output_unscaled,color="g",s=10)
pbmi.scatter(bmi, output_unscaled,color="b",s=10)
pbp.scatter(bp, output_unscaled,color="y",s=10)
```

BMI and BP seems to be a good indicator of disease progression.

```python
fig, gender = plt.subplots(layout='constrained', figsize=(20,2))
gender.scatter(output_unscaled, sex,s=15)
print(np.unique(sex, return_counts=True))
plt.show()
```

Showing Disease Progression for gender feature in a more visualized way. No pattern seems to emerge

```python
s1 = input_unscaled[:,4]
s2 = input_unscaled[:,5]
s3 = input_unscaled[:,6]
s4 = input_unscaled[:,7]
s5 = input_unscaled[:,8]
s6 = input_unscaled[:,9]
disease = output_unscaled
```

```python
fig, axs = plt.subplots(2, 3, layout='constrained', figsize=(14,7),)
ps1, ps2, ps3 = axs[0, 0], axs[0, 1], axs[0, 2]
ps4, ps5, ps6 = axs[1, 0], axs[1, 1], axs[1, 2]

ps1.set_xlabel("S1 (Total Serum Cholesterol)")
ps2.set_xlabel("S2 (Low-Density Lipoproteins)")
ps3.set_xlabel("S3 (High-Density Lipoproteins)")
ps4.set_xlabel("S4 (Total Cholesterol / HDL Ratio)")
ps5.set_xlabel("S5 (Log of Serum Triglycerides Level)")
ps6.set_xlabel("S6 (Blood Sugar Level)")

ps1.set_ylabel("Disease Progression")
ps2.set_ylabel("Disease Progression")
ps3.set_ylabel("Disease Progression")
ps4.set_ylabel("Disease Progression")
ps5.set_ylabel("Disease Progression")
ps6.set_ylabel("Disease Progression")

ps1.scatter(s1, output_unscaled, color="r",s=10)
ps2.scatter(s2, output_unscaled, color="g",s=10)
ps3.scatter(s3, output_unscaled, color="b",s=10)
ps4.scatter(s4, output_unscaled, color="y",s=10)
ps5.scatter(s5, output_unscaled, color="orange",s=10)
ps6.scatter(s6, output_unscaled, color="magenta",s=10)
```

```python
From these extra features, S3, S5 and S6 looks to be correlated to disease progression.
```
