```python
import sympy as sp
```

```python
i = sp.symbols('i', integer=True)

m = 4 # training examples
n = 2 # features

layer_counts = [1,1]

X = sp.MatrixSymbol('X', m, n)
Y = sp.MatrixSymbol('Y', m, 1)

A = sp.MatrixSymbol('A', n, layer_counts[0])
B = sp.MatrixSymbol('B', layer_counts[0], layer_counts[1])

print("X Y - ",X.shape,Y.shape)
print("A B - ",A.shape,B.shape)
```

```python
sigmoid = lambda x: 1/(1+sp.exp(-x))
```

```python
Z1 = X*A
L1 = Z1.applyfunc(sigmoid)
display(L1)
```

```python
Z2 = L1*B
P = L2 = Z2.applyfunc(sigmoid)
display(L2)
```

cross entropy

```python
cross_entropy_cost = sp.summation(-Y[i,0] * sp.log(P[i,0]) - (1 - Y[i,0]) * sp.log(1 - P[i,0]), (i,0,n))
cross_entropy_cost = sp.factor(cross_entropy_cost)
cross_entropy_cost = sp.cancel(cross_entropy_cost)
cross_entropy_cost = sp.simplify(cross_entropy_cost)
display(cross_entropy_cost)
print(sp.latex(cross_entropy_cost))
```

```python
CP = [cross_entropy_cost.diff(element) for element in B]
display((sp.factor(sp.simplify(sp.factor(CP[0])))))
```
