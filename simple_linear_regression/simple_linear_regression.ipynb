```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
```

```python
diabetes = load_diabetes(scaled=True) # enable Feature Scaling (std normalization)
data = diabetes.data

severity = diabetes.target # output of severity of diabetes or disease progression
bmi = data[:,2] # choose bmi as a starter feature
print(bmi.shape)
print(bmi)
```

```python
def mse_cost(w,b,x,y):
    m = x.shape[0] # number of training examples
    prediction = w * x + b
    error = prediction - y
    return np.dot(error, error) / (2*m)
```

```python
def compute_gradient(w,x,b,y):
    m = x.shape[0] # number of training examples
    prediction = w * x + b
    error = prediction - y
    dw = np.dot(error, x) / m
    db = np.sum(error) / m
    return dw, db
```

```python
w=0
b=0
# initialize model parameters

a=0.3
max_iterations=12000
iteration = np.empty(0, dtype=np.uint)
costs = np.empty(0)

for i in range(max_iterations):
    dw, db = compute_gradient(w,bmi,b,severity)

    w-=a*dw
    b-=a*db
    # update the weights
    
    iteration = np.append(iteration, i)
    costs = np.append(costs, mse_cost(w,b,bmi,severity))
    # add to array for visualization   
```

Cost Graph over Gradient Descent Iterations

```python
iteration = iteration[10:]
costs = costs[10:]

# graph the cost after updating the model
fig, cost_graph = plt.subplots(layout='constrained')

cost_graph.set_xlabel("Current Iteration")
cost_graph.set_ylabel("Cost")

cost_graph.plot(iteration, costs, color="r")
```

Simple Linear Regression Prediction

```python
fig, bmi_graph = plt.subplots(layout='constrained')

bmi_graph.set_xlabel("Body Mass Index")
bmi_graph.set_ylabel("Severity")

bmi_graph.scatter(bmi, severity, color="b", s=10)
bmi_graph.plot(bmi, w*bmi+b, color="g")
```

```python
sgd = SGDRegressor(max_iter=1000, tol=1e-3)
reg = make_pipeline(StandardScaler(), sgd)
bmi2d = bmi.reshape(-1, 1)
reg.fit(bmi2d,severity)
sgdout = reg.predict(bmi2d)
```

Prediction by SGDRegressor

```python
fig, bmi_graph = plt.subplots(layout='constrained')

bmi_graph.set_xlabel("Body Mass Index")
bmi_graph.set_ylabel("Severity")

bmi_graph.scatter(bmi, severity, color="b", s=10)
bmi_graph.plot(bmi, sgdout, color="g")
```
